# Creating Subdirectories/Folders to File Source Documents

```python
loopMinerals = minerals.drop_duplicates(subset = "LotNumber")

loopMinerals = loopMinerals.loc[loopMinerals['Plot (Y/N/Maybe)']=='Y']

loopMinerals.head()
```

```python
loopMinerals["Folder"] = ''

loopMinerals
```

## Create Directories

```python
for row in range(0,len(loopMinerals)):

    folderName = ''
    folderName = "Lot " + str(loopMinerals["LotNumber"].iloc[row]) + ' - '
    
    #names of lots are too long to save as folders, there for truncating the name by splitting and taking only partial elements
    splitName = loopMinerals["Name"].iloc[row].split(' - ')
    renameWord = splitName[0].split(" ")[0] + "-" + splitName[-1]
    
    #reconcatenating the folder name from the 2 splits
    folderName += renameWord
    
    #creating folder name at end of dataframe to move files to
    loopMinerals["Folder"].iloc[row] = folderName
    
    try:
        #creating directory for folders
        os.mkdir("Data Packages Source Files/"+folderName)
    except FileExistsError:
        print("File exists - continuing")
    
    print(folderName)
```

```python
def moveDownloads(PackageLinkFunctional = True ):
    
    '''
        This function will find downloaded files from the web and move them to their destination folder
        lotNumber will be a global variable in this script, which is generated and updated after each iteration of the parent loop
    '''
    #defining downloads directory to pull downloads from
    downloadDir = "/Users/Mishaun_Bhakta/Downloads/"
    downloads = os.listdir(downloadDir)

    #retrieving folder pertaining to lot 
    LotFolder = loopMinerals["Folder"].iloc[row]

    #initializing empty array to store downloaded file names
    filestomove = []
    
    #if the web function finds a package link that works, move the zip and index.pdf file
    #if it does not, then find pdf's in downloads folder and move them 
    if PackageLinkFunctional:
    
        #searching for zip file by looking for its lot number in the zip file name
        try:
            zipfile = list(filter(lambda x: str(lotNumber) in x, downloads))[0]
            filestomove.append(zipfile)
        except:
            print("error occured on finding zip file")

        #looking for the pdf file, the default download name is index.pdf from EnergyNet's database
        try:
            pdffile = list(filter(lambda x: "index" in x, downloads))[0]
            #creating name to rename the pdf file and using os package to rename the file it its respective directory
            renamedPdf = "Lot " + str(lotNumber) + " Full Package.pdf"
            os.rename(os.path.join(downloadDir, pdffile), os.path.join(downloadDir, renamedPdf))
            filestomove.append(renamedPdf)
        except:
            print("error occur occured on finding pdf file")
    else:
        try:
            #have to ensure downloads folder is empty prior to running code so that misc pdf's aren't moved
            pdffile = list(filter(lambda x: ".pdf" in x, downloads))
            for file in pdffile:
                filestomove.append(file)
        except:
            print("error occur occured on finding pdf file")
        

    #for each file in the filestomove, move them to its respective folder
    for file in filestomove:

        #moving files from downloads folder to respective lot package folder
        #using os.path.join to formulate directories cleanly
        dataPackagesDir = os.path.join(filepath, "Data Packages Source Files", LotFolder)
        shutil.move(os.path.join(downloadDir,file), dataPackagesDir)

   
```

```python
def searchLotWeb():
    #buyer's home page - have to navigate back to this spot each time
    authenticatedHomePage = 'https://www.energynet.com/bidder_home.pl?login=1'
    driver.get(authenticatedHomePage)
    time.sleep(3)
    

    #clicking load lots - this will reveal a search field to search by lot number
    driver.execute_script("$(\'#load_lots_3\').click()")
    time.sleep(3)

    #clicking the javascript generated search field
    searchField = driver.find_element_by_css_selector("#bidder_home_table_3_filter > label > input")
    searchField.click()
    time.sleep(1)

    #sending lot number of interest from pandas dataframe holding packages won by lot number
    searchField.send_keys(lotNumber)
    time.sleep(3) #allowing results to populate - entire database has 500 entries so takes a few seconds

    #finding package webpage by finding link text of  the lotNumber
    lotFound = driver.find_element_by_link_text(str(lotNumber))
    lotFound.click()
    time.sleep(5) #long sleep time because it takes a while for this page to load
```

```python
def downloadIndividualDocs():
    '''
        This function will download each file individually when there is no consolidated pdf hosted on EnergyNet
        for the package of interest
    
    '''
    #function to go to buyer's homepage, load won lots, and query by lot number
    searchLotWeb()
    
    #Locating "All Documents" button and clicking to get into data room
    try:
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "#idr > div > div > a:nth-child(1) > span")))
    except:
        time.sleep(10)
        print("Waited 10 additional seconds - could not find all documents button")
        
    allDocs = driver.find_element_by_css_selector("#idr > div > div > a:nth-child(1) > span")
    #allDocs.click()
    #selenium function will not work on this element, using javascript execution as work arround
    driver.execute_script("arguments[0].click();", allDocs)

    #locating all links with class page_name - this will allow us to loop through and click each one
    docDivs = driver.find_elements_by_class_name("page_name")

    for i in range(0,len(docDivs)):
        docDivs[i].click()
        
        #locating download button and clicking - brings up a javascrip genrated file download box
        #this box is not interactable with selenium - have to use JSexecute script 
        DLbutton = driver.find_element_by_id("docset_download")
        DLbutton.click()
        time.sleep(.5) #small sleep interval to ensure javacript generated popup box loads
        driver.execute_script("$(\"#download_options > div:nth-child(1) > a > img\").click()")
```

```python

#loggedInUrl

def automateWebDownloads():
    
    '''
        This function will navigate the EnergyNet Buyer's home page
        It will load the won lots with a javascript script
        It will then pass the lot number from the pandas dataframe holding mineral tracts we purchased
        Finally, it will navigate to the package data page and download the zip and pdf files pertaining to the package
        
        variables in function will depend on global variables defined in this script
            ie: driver, lotnumber
    
    '''
    #function to go to buyer's homepage, load won lots, and query by lot number
    searchLotWeb()

    ### *** In Order for pdf to be downloaded - change chrome privacy setting -> site settings -> pdf to automatically download
    ### This has to be set each time a new chromedriver session is launched
    
    try:
        #finding download links
        downloadPDF = driver.find_element_by_css_selector('#data-pack > div > ul > li:nth-child(1) > a')
        downloadPDF.click()

        downloadZIP = driver.find_element_by_css_selector('#data-pack > div > ul > li:nth-child(2) > a')
        downloadZIP.click()
        
        if type(None) != type(downloadPDF):
            #allows files to download before moving downloads
            time.sleep(7)
            
        try:
            moveDownloads()
        except:
            print( "Lot " + str(lotNumber) + "Files couldn't be moved")
            
        
    except:
        print("Lot " + str(lotNumber) + " - Full Consolidated Data packages may not have been hosted on EnergyNet")
    
        downloadIndividualDocs()
        time.sleep(20) #allows for the files to be fully downloaded before moving files
        #setting parameter of function to false to move all pdf's in downloads folder to destination folder
        try:
            moveDownloads(PackageLinkFunctional=False)
        except:
            print( "Lot " + str(lotNumber) + "Files couldn't be moved")
    
```

```python
for row in range(0,len(loopMinerals)):
    lotNumber = loopMinerals["LotNumber"].iloc[row]
    
    #assigining destinatino folder to check to see if there are pdf files in it
    destinationFolder = os.path.join("Data Packages Source Files", loopMinerals["Folder"].iloc[row])
  
    #This checks to see if the lot we are looking to pull files online have already been done.
    #if there are files existing in the folder, it will skip that lot number
    if len(os.listdir(destinationFolder))<1:
        automateWebDownloads()
    else:
        print("Files already exist in lot " + str(lotNumber) +"'s folder - skipping to next lot")
```
